```python
import pandas as pd
import re
```

```python
df_dados = pd.read_csv('dados.csv')
df_dados
```

```python
df_jogadores = pd.DataFrame()
df_time = pd.DataFrame()

seq = df_dados["Link"]
i=0

for link in seq:
    list_times = pd.read_html(link)   
    df_time = pd.concat([list_times[0],list_times[1]])
    df_time["CLUBE"] = df_dados["Clube"][i]
    df_jogadores = pd.concat([df_jogadores,df_time])
    i+=1
```

```python
df_jogadores.head(5)
```

```python
df_jogadores.reset_index(inplace=True)

df_jogadores = df_jogadores[(df_jogadores["J"] != "--")]

df_jogadores.drop(df_jogadores.columns.difference(["Nome","Idade","Alt","P",'J',"CLUBE"]),axis=1,inplace=True)

df_jogadores["Nome"] = df_jogadores["Nome"].apply(lambda x: re.sub('[0-9]','',x)) 

df_jogadores.rename({'Nome':'Apelido','CLUBE':'Clube','P':'Peso','J':'Jogos','Alt':'Altura'},axis=1, inplace=True)

df_jogadores
```

```python
df_jogadores.to_csv('jogadores_caracteristicas.csv',index = False)
```
